>**문자열 인덱싱**

문자열에서 글자(와 공백)는 각각 일련의 주소를 가진다. 우리는 특정 문자에 한하여 액세스하기 위하여 인덱싱을 사용한다.

기본적으로 파이썬에서는 첫번째 문자를 0으로 기준을 잡고, 뒷글자로 갈수록 점점 숫자가 커지는 형태를 취한다. 뿐만 아니라 문자열의 맨 끝자리부터 앞글자로 갈수록 -1부터 점점 작아지는 형태를 가진다.

image.png

'Hello' 라는 문자열이 있다고 치면

'H'는 인덱스 0

'e'는 인덱스 1

'l'는 인덱스 2

'l'는 인덱스 3

'o'는 인덱스 4

라고 할 수 있고, 각각의 인덱스를 불러오는 것은 아래를 통해 가능하다.


````python
text = "Hello"
for i in range(5):
  print(text[i]) #5번 반복하며 i에 순차적으로 0,1,2,3,4를 넣는다고 이해하면 된다.
````

문자열 인덱싱은 문자열에서 각각의 글자가 하나의 주소를 가지는 것을 의미했다.

그렇다면, 이 주소를 활용해서 특정 범위, 특정 규칙에 의해서 텍스트를 가져오는 방법은 무엇일까?

>**문자열 슬라이싱**

문자열의 주소 값(인덱싱)을 기반으로 문자열의 부분 값을 반환하는 기법이다.

문자열 슬라이싱은 기본적으로 아래와 같은 형태로 구현 가능하다.

````python
text[시작 인덱스(포함O):종료 인덱스(포함X):스텝(몇칸마다 탐색할 것인가, 선택)]
````



*   시작 인덱스: 몇번째 인덱스부터 반환할 것인가?
*   종료 인덱스: 몇번째 인덱스 전까지 반환할 것인가? *종료 인덱스에 해당하는 문자열은 포함되지 않음에 유의*
*   스텝(선택): 몇칸마다 탐색할 것인가? (기본값=1)



````python
#스텝은 선택이다. 물론 다른 것들도 선택이지만,
#스텝이 선택이라는건 종료인덱스 뒤에 들어가는 : 마저도 입력하지 않아도 된다는 의미이다.
#예시를 통해 이해해보자!

text='Hello'

print(text[0:6]) #범위를 초과하더라도 최대범위로 환산하여 'Hello'를 출력한다. 에러 없이!
print(text[-3:2]) #아무것도 출력되지 않는다. 아래 사진을 참고하여 왜 출력되지 않았을지 생각해보자. 답은 >>>>>>>>>> -3인덱스가 Hello의 세번째 글자인 l에 해당하고, 2인덱스 또한 세번째 글자에 해당하는데, 슬라이싱 특성상 [a:b]에서 a를 포함하고, b를 포함하지 않기 때문.
print(text[::2]) #앞 두개의 값에 아무것도 입력하지 않아도 된다.

print('엄마가준비한식사'[::3]) #엄준식이다!
````

>**split()**

split 함수는 말 그대로 텍스트를 **특정 값을 기준으로** 분리하여 **리스트**의 자료형으로 반환하는 것이다. *리스트는 뒷장에서 다루고 있다.*

'text'.split('텍스트를 나누는 기준이 될 문자') 로 구성된다.

````python
text='안녕하세요엄저의엄이름은엄태영준엄입니다'
print(text.split('엄'))

text='인하대학교 데이터사이언스학과는 데이터 분석, 인공지능, 빅데이터 등의 분야에서 전문가를 양성하는 학과이다.'
print(text.split()) #중요: split의 입력값에 아무것도 넣지 않는 경우, 띄어쓰기를 기준으로 분리한다. (기본값)
````

````python
#언패킹
items='umm jun sik'.split()
print(items)
a,b,c=items #items 리스트의 각 값을 a,b,c 라는 변수로 "언패킹"하는 과정이다.
print(a)
print(b)
print(c)
````

>**Formatting**



파이썬에서 %s 같은 표현은 문자열 포매팅에서 사용되는 형식 지정자(format specifier) 중 하나로, 일반적으로 특정 형식으로 값을 출력할 때 사용된다. % 기호는 포매팅 연산자이다.

흔히 사용하는 것은 %d, %s, %f 등이다:

>**조건문**

조건문의 구조는 아래와 같다.
```
if <조건>:
  <조건 충족시 수행할 명령>
else:
  <조건 비충족시 수행 명령>
```
파이썬의 IF문은 조건 뒤에 콜론을 삽입하여야 한다는 특징이 있다.

들여쓰기(Tab키)를 사용해 조건 충족/비충족 시에 실행할 코드를 구분할 수 있다.

else: 는 안 써도 된다. 만약 안 쓰게 되면 조건이 충족될 때에만 코드가 실행되는 것.



````python
#IF
score = int(input("점수를 입력하세요:  ")) #사용자에게 점수를 입력받고 이를 정수형으로 변환
if score >= 60: #만약 점수가 60점 이상인 경우에는 'score >= 60' 이라는 코드는 True의 값을 가진다. 아닐 경우에는 False의 값을 가진다.
  print("합격입니다.") #60점 이상인 경우
else:
  print("불합격입니다.") #60점 미만인 경우
````

참고: input()함수는 입력값이 무엇이든지간에 입력값을 str 형식으로 저장한다. 그래서 우리는 정수형을 입력받을 때 `int(input())`을 사용한다.

그런데 input값이 정수의 형태가 아니라면 (예: a) ValueError가 발생한다.

따라서, 만약 사용자가 정수가 아닌 값을 입력했을 때, 예외 처리의 방법을 고안하여야 한다.

더 자세한 내용을 공부하고자 한다면 try-except문이나 isdigit()을 찾아보도록 하자.

````python
#cf) 해결방법 - int를 입력받을 때까지 계속 반복. while문과 조합하여 구현 가능하다.
#
while True:
  score = input("점수를 입력하세요:  ") #사용자에게 점수를 입력받고 이를 정수형으로 변환
  if score.isdigit(): #score.isdigit() == True << 가 생략 (해도됨)
    score = int(score)
    if score >= 60: #만약 점수가 60점 이상인 경우에는 'score >= 60' 이라는 코드는 True의 값을 가진다. 아닐 경우에는 False의 값을 가진다.
      print("합격입니다.") #60점 이상인 경우
    else:
      print("불합격입니다.") #60점 미만인 경우
    break
  else:
    print("정수만 입력할 수 있습니다. 다시 시도하세요.")
````

>**while문 (조건 제어 반복문)**

while 뒤의 조건을 충족하는 경우 실행되며, **해당 조건이 충족되지 않을 때까지 실행된다.**


아래는 완전히 같은 결과를 만드는 반복문 세개이다.

````python
i = 0
while i < 10: #i가 10보다 작을 때까지만 반복. 만약 10 이상이 되면 반복이 중단되고, 만일 다시 i가 10보다 작아지더라도 while문을 별도로 실행시키지 않는 한 다시 실행되지 않고 지나간다. 이것은 interpreter 언어의 기본적 원리이기도 하다.
  print(i)
  i += 1 #i = i+1 과 동치. 파이썬의 편의 기능 중 하나이다.
````

````python
#참고: for문과 while문은 상호 변환 가능하다. 반복을 판단하는 방식이 다를 뿐이다.
#아래는 위 블럭을 for 문으로 변환한 것이다.
for i in range(10):
  print(i)
````

````python
#while문 제어: 반복문(while,for)내의 break 는 반복문을 종료시킨다.
i=0
while True:
  print(i)
  i += 1
  if i >= 10:
    break
````

**continue**

반복문 전체를 빠져나오는 것이 아니라 해당 조건만 건너뛰고 싶을 때는 어떻게 하면 될까? 예를 들어, 화면에 1부터 10까지 출력하는데 5인 경우에만 출력하지 않고 건너뛰는 것처럼 말이다. 이럴 때 사용할 수 있는 키워드가 바로 continue이다.

````python
num = 0
while num < 10:
  num += 1
  if num == 5:
    continue
  print(num)
````

객체지향 프로그래밍의 개념을 배우고 이를 적용하는 클래스의 개념을 배웁니다.

>**1. 객체지향 프로그래밍의 이해**

객체지향 프로그래밍(Object-Oriented Programming, OOP)은 데이터를 중심으로 프로그램을 구조화하는 방식이다. OOP에서는 **객체**라는 개념을 사용하여 데이터와 기능을 하나의 단위로 묶는다. 현실 세계의 사물처럼 프로그램 내에서 독립적으로 동작하는 객체들이 상호작용하며 프로그램이 실행된다.

예를 들어, 자동차를 객체로 본다면, 자동차는 여러 가지 **속성(데이터)**, 즉 색상, 모델, 속도 등을 가지고 있으며, **행동(메서드)**, 즉 가속, 브레이크와 같은 동작을 수행할 수 있다.

**그래서 이것은 그동안의 코딩과 어떤 점이 다른가?** 라는 의문이 들 수도 있다.

객체지향 프로그래밍(OOP)은 데이터와 그 데이터를 처리하는 동작을 하나의 단위(객체)로 묶는 방식이다. 반면에, 기존의 절차지향 프로그래밍은 코드가 순차적으로 실행되는 방식으로, 동작을 중심으로 코드를 작성한다는 차이가 있다.

- 절차지향에서는 데이터를 다루는 함수들이 따로 존재하고, 데이터가 필요할 때 함수가 그 데이터를 가져와서 처리한다.

- 객체지향에서는 데이터와 그 데이터를 처리하는 메서드가 객체라는 단위 안에 함께 묶여있어, 각 객체가 자신만의 데이터를 관리하며 필요한 동작을 수행한다.

또한 객체 지향 프로그래밍의 특징으로는 **캡슐화, 상속, 다형성**이 있다. (수업자료)

- 가시성 : 객체의 정보 접근을 숨기는 것
- 상속 : 부모 클래스에 정의된 변수/함수를 자식 클래스가 물려 받는 것
- 다형성 : 같은 이름의 메서드가 다른 기능을 할 수 있도록 하는 것

>**2. 클래스(Class)란 무엇인가**

**클래스(Class)**는 객체를 생성하기 위한 청사진(설계도)이다. 클래스는 동일한 종류의 객체들을 생성할 수 있게 해준다. 예를 들어, '자동차'라는 클래스를 정의하면, 그 클래스를 통해 여러 대의 자동차 객체를 만들 수 있다. 각 자동차는 서로 다른 속성(색상, 모델 등)을 가질 수 있지만, 동일한 구조(메서드와 속성)를 공유한다.

다른 예시로, 수업 시간에 나왔던 '붕어빵 틀'을 들 수 있다. 예를 들어, '붕어빵틀' 이라는 클래스를 정의하면, 우리는 그 클래스를 통해 여러 개의, 다양한 속성(붕어빵에서는 팥붕/슈붕이 될 것이다)을 가진 **서로 다른 객체인 붕어빵**을 만들어낼 수 있다.

클래스를 사용하는 이유는 프로그램의 구조를 명확하게 하고, 코드의 재사용성을 높이며, 유지보수를 쉽게 하기 위함이다.

*붕어빵틀 없이 붕어빵을 만든다고 생각해 보자. 붕어빵은 과연 천원대로 구매할 수 있는 간식이겠는가? 붕어빵틀이 붕어빵을 합리적인 가격에 먹을 수 있는 도구가 된다는 관점에서 클래스도 마찬가지이다.*

>**3. 속성(데이터)과 행동(메서드)**

클래스는 두 가지 주요 요소로 구성된다: **속성과 메서드**

- 속성(데이터): 객체가 가지고 있는 특성이나 상태를 나타낸다. 예를 들어, 자동차의 색상이나 모델은 속성이다.

- 메서드(행동): 객체가 수행할 수 있는 동작을 정의한다. 자동차가 가속하거나 멈추는 동작은 메서드이다.

지금부터 속성은 계속 속성으로, 행동은 메서드로 부를 것이다.

>**4. 클래스는 어떻게 구성되는가?**

클래스는 주로 두 가지 요소로 구성된다: 속성(데이터)을 정의하는 생성자 메서드와 행동(메서드)이다.

**속성: `__init__` 메서드**

`__init__` 메서드는 객체가 생성될 때 자동으로 호출되는 생성자(constructor) 메서드이다. 객체가 생성될 때 초기 값을 설정해주는 역할을 한다. 이 메서드 안에서 클래스의 속성들이 정의된다.

init 메서드는 굳이 앞뒤로 언더스코어 두개(__) 를 두르고 있는데, 그 이유는 이 메서드가 파이썬에서 미리 특정 역할을 하기로 정해놓은 특수 메서드(magic method)이기 때문이다. 해당 내용은 [7장] 에서 알아본다.

예시:
```python
class Car:
    def __init__(self, color, model):
        self.color = color  # 속성 정의 (색)
        self.model = model  # 속성 정의 (모델)
        self.speed = 0 #속도의 초기값은 0으로 하자! (차가 출고할때부터 달리고 있진 않다)
```

**행동: 메서드**

메서드는 객체의 행동을 정의한다. 메서드는 클래스 내부에 정의된 함수로, 객체가 어떤 동작을 수행할지를 결정한다.

예시:
```python
    def accelerate(self):  # 가속 메서드
        self.speed = self.speed + 10 #가속 메서드를 한번 호출하면 객체의 속도를 +10km/h 만큼 변화시킨다.
        print(f"{self.model}가 {self.speed}km/h로 주행 중입니다.")
    
    def brake(self):  # 감속 메서드
        self.speed = self.speed - 10 #감속 메서드를 한번 호출하면 객체의 속도를 -10km/h 만큼 변화시킨다.
        if self.speed > 0 :
          print(f"{self.model}가 {self.speed}km/h로 주행 중입니다.")
        else:
          print(f"{self.model}가 멈췄습니다.")
```
클래스는 일반적으로 객체를 만들 때 설정할 **속성**과 속성을 수정할 수 있는 방법인 **메서드**를 같이 정의한다. 위에서 본 각각의 속성과 메서드들을 모두 합치면 하나의 **클래스**가 완성된다.

완성본을 한번 보자:

```python
class Car:
```

>**6. 왜 `self`가 필요한가?**

클래스 내에서 메서드를 호출할 때, Python은 그 메서드를 호출한 객체가 누구인지를 알아야 한다. 예를 들어, Dog 클래스에서 여러 개의 객체가 생성될 수 있는데, 그 각각의 객체는 독립적으로 다른 속성을 가질 수 있다. 이를 구분하기 위해 self가 필요하다.

코드
```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name}가 짖습니다.")

dog1 = Dog("Hamo", 11)
dog2 = Dog("Coco", 9)

print(dog1.bark)
print(dog2.bark)
```

결과
```python
Hamo가 짖습니다.
Coco가 짖습니다.
```

위 코드에서 self.name과 self.age는 해당 객체의 name과 age 속성을 가리킨다.

만약 `dog1.bark`가 호출되면, 이 메서드를 호출한 객체는 `dog1`이므로 class 안의 bark(self) 함수에서 `self`가 `dog1`로 치환된다고 이해해도 좋다.

만약 self가 없다면 어떠하겠는가? 클래스 내의 메서드는 어떤 객체가 자신을 호출했는지 알지 못해 호출한 객체에 맞는 값을 내놓을 수 없다.

self는 반드시 첫 번째 매개변수로 지정되어야 하지만, 메서드를 호출할 때는 명시적으로 전달하지 않는다. Python이 내부적으로 자동으로 전달한다. (우리가 `dog1.bark(dog1)`라 하지 않고 `dog1.bark`만 해도 되는 이유)

>**7. 특수 메서드(magic method)**

특수 메소드는 magic method, special method, double underscore, dunder method 등 다양하게 불리며, 특수 상황에서 자동으로 호출되는 메소드를 말한다.
특수 메소드의 종류는 여러가지가 있는데, 해당 포스팅에서는 객체 지향 프로그래밍을 위해서 꼭 알아야하는 __init__메소드와 __str__메소드 두가지만 다루어본다.


- `__init__` 메소드 [이닛 메소드]

  인스턴스가 생성될 때 자동으로 호출하는 메소드
  __init__메소드를 사용하면 코드가 훨씬 간결해짐을 느낄 수 있다.

- `__str__` 메소드 [던더 에스티알]

  print 함수를 사용할 때, 자동으로 호출
  인스턴스를 출력할 때 우리가 원하는 정보를 print하고 싶을 때 사용

  그러나 `__str__` 메서드를 정의하면, 객체를 출력하려고 했을 때 `__str__`메서드 안의 return 값이 출력됨.

>**8. 예시를 통한 이해**

몇가지 예시를 통해 이해하여 보자.

````python
class SoccerPlayer(object):
    def __init__(self, name, position, back_number):
        # self.name은 이 객체의 name 속성을 나타냄
        # 생성자가 호출될 때 전달된 name 값을 객체의 name 속성에 저장함
        self.name = name

        # self.position은 객체의 position 속성
        # 전달된 position 값을 객체의 position 속성에 저장함
        self.position = position

        # self.back_number는 객체의 등번호를 저장하는 속성
        # 전달된 등번호 값을 객체의 back_number 속성에 저장함
        self.back_number = back_number

# SoccerPlayer 클래스(붕어빵틀)을 사용해 messi 라는 인스턴스(객체)를 생성
# self는 따로 쓰지 않아도 자동으로 처리되고, "Lionel Messi", "MF", 10이 각각 name, position, back_number로 전달됨
# self 관련 내용은 하단을 참조
messi = SoccerPlayer("Lionel Messi", "MF", 10)

# messi 객체의 속성(name, position, back_number)을 이용해 메시지 출력
print(f"{messi.name}은 {messi.position} 포지션이고, 등번호는 {messi.back_number}입니다.")

# messi 객체 자체를 출력하려고 하면, 기본적으로 객체의 메모리 주소가 출력됨
print(messi)
````

위의 코드에서는 클래스 안에서 메서드(함수)를 정의하지 않아서 name, position, back_number 숫자를 바꿀 수 있는 기능이 존재하지 않는다.

물론, `messi.name = 'christiano ronaldo'` 처럼 속성(변수)를 다시 정의해서 바꿀 순 있지만, 클래스 내에 해당 기능이 존재하지 않는다는 것이다.

따라서 클래스를 정의할 때, 속성(name, position, back_number)과 함께 def를 통해 메서드(함수)를 정의해주어 속성(name, position, back_number; 변수)을 바꿀 수 있는 기능을 내장함이 좋다.

아래 내용을 통해 제대로 이해하여 보자:

````python
class SoccerPlayer(object):
    def __init__(self, name, position, back_number):
        # init을 통해 객체의 속성 name, position, back_number을 할당
        self.name = name
        self.position = position
        self.back_number = back_number

    def change_back_number(self, new_number):
        # 선수의 등번호를 변경하는 메서드
        print("선수의 등번호를 변경합니다: From %d to %d" % (self.back_number, new_number))
        self.back_number = new_number  # 새로운 번호로 등번호를 업데이트함

    def exchange_back_number(self, target_player):
        # 선수의 등번호를 입력된 다른 선수와 바꾸는 메서드
        print("%s 선수의 등번호를 %s와 변경합니다: From %d to %d" % (self.name, target_player.name, self.back_number, target_player.back_number))
        temp = target_player.back_number
        target_player.back_number = self.back_number
        self.back_number = temp

    def __str__(self):
        # messi 객체 자체를 출력하려고 하면, 기본적으로 객체의 메모리 주소가 출력됨
        # 그러나 __str__ 을 정의하면, 객체를 출력하려고 했을 때 __str__메서드 안의 return 값이 출력됨
        return "안녕하세요, 저는 축구선수 %s 입니다. 저는 %s 포지션을 맡고 있습니다. 제 등번호는 %d번 입니다." % (self.name, self.position, self.back_number)


# messi 인스턴스(객체) 생성, 이름은 Lionel Messi, 포지션은 MF, 초기 등번호는 10번
messi = SoccerPlayer("Lionel Messi", "MF", 10)

# messi의 등번호를 10번에서 7번으로 변경
messi.change_back_number(7)

# messi 인스턴스(객체)를 출력, 이때 __str__ 메서드에 정의된 문자열이 출력됨
print(messi)

# jr_messi 인스턴스(객체) 생성, 이름은 Lionel Jr. Messi, 포지션은 MF, 초기 등번호는 10번
jr_messi = SoccerPlayer("Lionel Jr. Messi", "MF", 5)

# exchange_back_number 메서드를 사용하여 jr_messi와 messi 인스턴스의 back_number을 서로 바꿈
# 참고로, 이 코드를 실행하기 전 messi.back_number = 7, jr_messi.back_number = 5 이다.
jr_messi.exchange_back_number(messi)

# 어떻게 변했는지 보자.
print("\n어떻게 변했는지 보자!")
print(messi)
print(jr_messi)
````
